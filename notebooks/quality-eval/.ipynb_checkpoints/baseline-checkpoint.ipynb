```python
import json

# results_path = '/Users/lan/Documents/hpi/projects/aggregation-detection/code/aggregation-detection/aggrdet/baseline/baseline-results.jl'
results_path = '/Users/lan/Documents/hpi/projects/aggregation-detection/code/baseline/timeout-10mins/baseline-results.jl'

with open(results_path, 'r') as reader:
    result_json_dicts = [json.loads(line) for line in reader]
```

```python
result_json_dicts[0].keys()
```

# How many files on which the baseline fail to finish within the timout?

```python
success = []
failure = []
for json_dict in result_json_dicts:
    try:
        exec_time = json_dict['exec_time']['Baseline']
    except RuntimeError:
        print('Cannot load the exec time for file: %s' % json_dict['file_name'])
    else:
        if exec_time == -1:
            failure.append(json_dict)
        else:
            success.append(json_dict)

print('# success: %d' % len(success))
print('# failure: %d' % len(failure))
```

# Precision - Recall per file

```python
l = []
for json_dict in success:
    file_name = json_dict['file_name']
    sheet_name = json_dict['table_id']
    groundtruth = json_dict['aggregation_annotations']
    detected_aggregations = json_dict['detected_aggregations']
    exec_time = json_dict['exec_time']['Baseline']
    l.append([file_name, sheet_name, len(groundtruth), len(detected_aggregations), exec_time])
    
import pandas

pandas.set_option("display.max_rows", None)

quailty_df = pandas.DataFrame(data=l, columns=['file name', 'sheet name', '# aggregations', '# detected aggregations', 'exec time'])

quailty_df
```
